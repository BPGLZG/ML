# Proyecto Machine Learning

## Tema: Predicción de condiciones metabólicas a partir del microbioma intestinal humano

![imagen](image.jpg)

### Exploración de perfiles metagenómicos para identificar patrones asociados a salud y enfermedad

#### 1. Descripción:

El microbioma intestinal está relacionado con múltiples procesos de salud humana, incluyendo metabolismo, inflamación y enfermedad. Su composición puede variar entre individuos sanos y personas con ciertas condiciones clínicas.
<br>
* **Objetivo**: explorar si el microbioma contiene señal útil para diferenciar estados clínicos.

#### 2. Dataset elegido

##### Datos usados:

**Human Metagenomics** (https://www.kaggle.com/datasets/antaresnyc/human-metagenomics)
<br>
Datos y metadatos de más de 3000 metagenomas disponibles públicamente.
<br>
Este conjunto de datos fue creado por el equipo de Edoardo Pasolli, Duy Tin Truong, Faizan Malik, Levi Waldron y Nicola Segata; publicaron un artículo de investigación en julio de 2016 y crearon *MetAML*, un análisis de predicción metagenómica basado en aprendizaje automático.

Abundancias relativas a nivel de especie.
<br>
`Rows` casos (1989), `columns` - metadata (211) y especies (2128)

##### Datos después de la limpieza:

- 1577 muestras
- 2135 columnas
- 2128 variables taxonómicas
- 7 columnas de metadata

#### 3. Problema planteado:

Evaluar si los perfiles del microbioma intestinal permiten identificar patrones asociados a condiciones clínicas metabólicas.
<br>
- <u>**Target candidato principal**</u>: perfiles sanos *vs*. perfiles con diabetes tipo 2
<br>
- <u>Target secundario</u>: perfiles sanos *vs*. perfiles con obesidad

#### 4. Primeros hallazgos del EDA:

1. El dataset estaba duplicado por sampleID y se limpió.
2. Hay desbalance en disease.
3. Hay alta sparsity en taxones.
4. PCA inicial muestra que la variabilidad está distribuida en muchas dimensiones.
    * PC1: 5.5%
    * PC2: 2.51%
    * **Total: 8.03%**
        <br>
        Es esperable en microbioma; PCA se usará como herramienta exploratoria, no como prueba definitiva.

#### 5. Riesgos metodológicos:

- Alta dimensionalidad.
- Muchas variables sparse.
- Clases desbalanceadas.
- Posible sesgo por dataset de origen, país o cohorte.
- La etiqueta n puede representar control del estudio, no necesariamente salud perfecta.

#### 6. Modelos a analizar:
- **Supervisados**:
    - Logistic Regression
    - Random Forest
    - Gradient boosting
    - SVM
    - KNN
- **No supervisados**:
    - KMeans


#### 7. Valor de negocio
- Equipo de investigación clínica
- Startup biotech/nutrición personalizada
<br>

**NOTA**: No sustituye diagnóstico médico; apoya análisis exploratorio y generación de hipótesis.
<br>
<u>La idea no es construir una herramienta de diagnostico médico</u>, sino un **modelo exploratorio** que permita detectar señales biológicas, comparar perfiles microbianos y evaluar si el microbioma puede apoyar tareas de investigación, cribado preliminar o priorización de muestras.
