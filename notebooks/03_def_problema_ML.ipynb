# Proyecto Machine Learning

## Tema: Predicción de condiciones metabólicas a partir del microbioma intestinal humano

![imagen](image.jpg)

### Definición del problema de Machine Learning

#### 1. Objetivo principal

**Clasificación binaria: n vs t2d**
<br>
¿Puede el perfil taxonómico del microbioma intestinal diferenciar controles (`n`) de muestras asociadas a diabetes tipo 2 (`t2d`)?

Por qué este `target`:

* Tiene suficientes muestras: `n` = 710, `t2d` = 223.
* Es clínicamente interpretable.
* Encaja con el enfoque metabólico.
* Evita empezar con una multiclase demasiado ruidosa.
* Permite explicar negocio/investigación con claridad.

#### 2. Features

**Modelo inicial:**
<br>
`X` = columnas taxonómicas k__
<br>
`y` = disease transformado a binario

#### 3. Modelos candidatos

**Supervisados**
1. Logistic Regression
2. Random Forest
3. Gradient Boosting
4. SVM
5. KNN

**No supervisados**
1. PCA + KMeans

#### 4. Métricas

- Accuracy
- Precision
- Recall (para t2d)
- f1-score*
- Roc_auc*
- Confusion matrix
